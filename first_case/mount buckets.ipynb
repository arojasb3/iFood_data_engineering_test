```python
a_key = dbutils.secrets.get('s3_db','a_key')
s_a_key = dbutils.secrets.get('s3_db','sec_a_key')
```

```python
encoded_secret_key = s_a_key.replace("/", "%2F")
aws_bucket_name = "ifood-de-test-arb"
mount_name = "ifood-de-test-arb"

dbutils.fs.mount("s3a://%s:%s@%s" % (a_key, encoded_secret_key, aws_bucket_name), "/mnt/%s" % mount_name)
display(dbutils.fs.ls("/mnt/%s" % mount_name))
```

```python
#dbutils.fs.mount("s3a://ifood-de-test-arb", "/mnt/ifood-de-test-arb")
dbutils.fs.unmount("/mnt/ifood-de-test-arb")
#display(dbutils.fs.ls("/mnt/ifood-de-test-arb"))
```
